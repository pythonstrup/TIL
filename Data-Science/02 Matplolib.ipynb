### 2-1. 간단한 선 그래프 그리기

<hr/>


````python
import matplotlib.pyplot as plt

# notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해줌
%pylab inline

def korean_population_chart():
    years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
    population = [2051, 2501, 3224, 3812, 4286, 4700, 4887, 4932]
    
    # x축 = years, y축 = population
    plt.plot(years, population, color='blue', marker='o', linestyle='solid')
    
    # 제목설정
    plt.title("Korean Population")
    
    # 레이블 추가
    plt.xlabel("Year")
    plt.ylabel("10 thousand of")
    plt.show()

korean_population_chart()
````

### 2-2. 막대 그래프 그리기

<hr/>

막대 그래프는 이산적인 항목들에 대한 변화를 보여줄 때 사용한다. 아래의 예시를 살펴보자.

````python
sports = ["basketball", "football", "hockey", "tennis"]
num_students = [10, 20, 5, 8]

plt.bar(range(len(sports)), num_students)

plt.title("Favorite Sports")
plt.xlabel("Sports")
plt.ylabel("# of Students")

# x축 막대에 긱 스포츠 종목 라벨 추가
plt.xticks(range(len(sports)), sports)
plt.show()
````

### 2-3. 히스토그램 그리기

<hr/>

히스토그램이란 표로 되어 있는 도수 분포를 정보 그림으로 나타낸 것이다. 즉, 정해진 구간에 해당하는 항목의 개수를 보여줘 값의 분포를 파악할 수 있는 그래프이다.

````python
from collections import Counter

heights = [173, 175, 162, 160, 180, 181, 159, 177, 195, 165, 168, 172, 179, 211]

# 최대키는 200으로 설정
histogram = Counter(min(height//10 * 10, 200) for height in heights)
histogram

plt.bar([x + 5 for x in histogram.keys()],
       histogram.values(),
       10,
       edgecolor=(0, 0, 0))

# x축은 145~215, y축은 0~10
plt.axis([145, 215, 0, 10])

# 계급 150부터 200까지 10단위로
plt.xticks([10 * i for i in range(15, 21)])
plt.xlabel("Height")
plt.ylabel("# of Students")
plt.title("Students's Height")
plt.show()
````

### 2-4. 여러 개의 선 크래프

<hr/>

여러 개의 선 그래프를 겹쳐 그릴 수 있다.

````python
arr = [1,2,4,8,16,32,64,128,256]
opposition = [256,128,64,32,16,8,4,2,1]
total_error = [x + y for x, y in zip(arr, opposition)]
xpos = [i for i, _ in enumerate(arr)]

plt.plot(xpos, arr, 'b-', label="Default Value")
plt.plot(xpos, opposition, 'r-.', label="Opposition")
plt.plot(xpos, total_error, 'b:', label="Total Error")

plt.title("Line Graph")

# lengend(loc=x)에서 loc의 범위는 0~10이다. 
# 값에 따라 선 정보에 대한 박스의 위치가 달라진다.
plt.legend(loc=0)
plt.xlabel("X position")
plt.ylabel("Y position")

# x 좌표 눈금 없애기
plt.xticks([])
plt.show()
````

#### 보너스
선 그래프를 그릴 때 선의 색상이나 스타일을 설정할 수 있다. 아래의 예시를 살펴보자.

````python
import numpy as np
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
````

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

````python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

### 2-5. 산점도

<hr/>

산점도(Scatterplot)은 두 변수 간의 연관 관계를 보여주는 그래프이다. 

````python
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n']
heights = [175, 152, 168, 180, 185, 176, 172, 163, 169, 177, 181, 171, 158]
weights = [65, 49, 55, 75, 82, 76, 69, 60, 58, 59, 90, 88, 55]

plt.scatter(weights, heights)

for label, weight, height in zip(labels, weights, heights):
    # 그래프에 주석달기
    plt.annotate(label,
                xy = (weight, height),     # text를 표시할 좌표
                xytext=(5, -5),            # 포인터에서부터의 라벨링 위치
                textcoords='offset points')# xy(좌표 축의 값)에서부터 xytext offset 위치에 출력

plt.title("Height & Weight")
plt.ylabel("Height(cm)")
plt.xlabel("Weight(kg)")
plt.show()
````
