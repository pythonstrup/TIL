### 4-1.  Hierarchical Index

<hr/>

- Groupby

SQL의 groupby 명령어와 같다.

```python
import pandas as pd

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df

df.groupby("Team")["Points"].sum()
```

```python
# 1개 이상의 column을 묶을 수도 있다.
# 연도별 팀 포인트
df.groupby(["Team", "Year"])["Points"].sum()
```

<br/>

- index

Groupby 명령의 결과물도 결국은 DataFrame이다. 두 개의 column으로 groupby를 할 경우, index가 2개가 생성된다.

```python
h_index = df.groupby(["Team", "Year"])["Points"].sum()
h_index.index
```

```python
h_index["Devils": "Kings"]
```

<br/>

- unstack

Group으로 묶여진 데이터를 matrix 형태로 전환해줌

```python
h_index.unstack()
```

<br/>

- swaplevel

index level을 변경할 수도 있다.

```python
h_index.swaplevel()
```

```python
h_index.sum(level=0)
```

### 4-2  GroupBy 메소드

<hr/>

- grouped

Groupby에 의해 Split된 상태를 추출 가능함. Tuple 형태로 그룹의 Key값과  Value 값이 추출된다.<br/>

```python
grouped = df.groupby("Team")

for name, group in grouped:
    print(name)
    print(group)
    print()
```

```python
grouped.get_group("Devils")
```

<br/>
추출된 group 정보에는 세 가지 유형의 apply가 가능하다.<br/>

1. Aggregation: 요약된 통계정보를 추출해줌
2. Transformation: 해당 정보를 변환해줌
3. Filtration: 특정 정보를 제거하여 보여주는 필터링 기능

<br/>

#### 1. agg 메소드

```python
grouped.agg(sum)
```

```python
import numpy as np
grouped.agg(np.mean)
```

```python
# 특정 컬럼에 여러 개의 function을 Apply할 수도 있음
grouped["Points"].agg([np.sum, np.mean, np.std])
```

<br/>

#### 2. transfrom 메소드

Aggregation과 달리 key값별로 요약된 정보가 아니라 개별 데이터의 변환을 지원하는 메소드이다.

```python
score = lambda x: (x)
grouped.transform(score)
```

```python
score = lambda x: (x - x.mean()) / x.std()
grouped.transform(score)
```

<br/>

#### 3. filter 메소드

특정 조건으로 데이터를 검색할 때 사용

```python
df.groupby("Team").filter(lambda x: len(x) >= 3)
```

```python
df.groupby("Team").filter(lambda x: x["Points"].max() > 800)
```

### 4-3  Pivot table Crosstab

<hr/>

#### Pivot Table

```python
import pandas as pd
import dateutil

df_phone = pd.read_csv("data/phone_data.csv")
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.pivot_table(["duration"],
                    index=[df_phone.month, df_phone.item],
                    columns=df_phone.network, aggfunc="sum", fill_value=0)
```

<br/>

#### Crosstab

특히, 두 컬럼에 교차 빈도, 비율, 덧셈 등을 구할 때 많이 사용한다. Pivot Table의 특수한 형태라고 생각하면 된다. User-Item Rating Matrix 등을 만들 때 사용가능하다.

```python
df_movie = pd.read_csv("data/movie_rating.csv")
df_movie.head()
```

```python
pd.crosstab(index=df_movie.critic, 
            columns=df_movie.title, 
            values=df_movie.rating,
            aggfunc="first").fillna(0)
```

```python
# Pivot Table로도 가능하다

df_movie.pivot_table(["rating"], 
                     index=df_movie.critic, 
                     columns=df_movie.title,
                     aggfunc="sum", fill_value=0)
```

### 4-4  Merge & Concat

<hr/>

#### Merge

SQL에서 많이 사용하는 Merge와 같은 기능이다. 두 개의 데이터를 하나로 합친다.

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
pd.merge(df_a, df_b, on='subject_id')
```

```python
pd.merge(df_a, df_b, left_on='subject_id', right_on='subject_id')
```

<br/>

- Join Method



```python
pd.merge(df_a, df_b, on='subject_id', how='left')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='right')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='outer')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='inner')
```

<br/>

- Index based join

사용할 일이 많지 않음

```python
pd.merge(df_a, df_b, right_index=True, left_index=True)
```

<br/>

#### Concat

같은 형태의 데이터를 붙이는 연산작업

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
df_new = pd.concat([df_a, df_b])
df_new.reset_index()
```

```python
df_a.append(df_b)
```

```python
df_new = pd.concat([df_a, df_b], axis=1)
df_new.reset_index()
```
