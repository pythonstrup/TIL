### 5-1. 기초

<hr/>

- pyplot 객체를 사용하여 데이터를 표시하는 시각화 도구이다.
- pyplot 객체에 그래프들을 쌓은 다음 show로 flush

<br/>

````python
import matplotlib.pyplot as plt

x = range(100)
y = [value**2 for value in x]
plt.plot(x, y)
plt.show()
````

<br/>

matplotlib의 최대 단점 argument를 kwargs로 받는다는 것이다. 고정된 argument가 없어서 alt+tab으로 확인이 어렵다.

```python
plt.plot(*args, *kwargs)
```

<br/>

Graph는 figure(pyplot의 일부분) 객체에서 생성된다. pyplot 객체 사용 시, 기본 figure에 그래프가 그려진다.<br/>
show() 메소드를 사용해서 flush가 일어난다.
 

````python
import numpy as np

x1 = range(100)
y1 = [np.cos(value) for value in x1]

x2 = range(100)
y2 = [np.sin(value) for value in x2]

plt.plot(range(100), range(100))
plt.plot(x1, y1)
plt.plot(y2, x2)
plt.show()
````

<br/>

### 5-2 figure

<hr/>

- 여러 개의 그림판으로 그래프 출력

````python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,10) # 크기지정
ax_1 = fig.add_subplot(1,2,1) # 두개의 plot 생성
ax_2 = fig.add_subplot(1,2,2)  # 두개의 plot 생성

ax_1.plot(x1, y1, c="b")  # 첫번째 plot
ax_2.plot(x2, y2, c="g")  # 두번째 plot
plt.show() # show & flush
````

<br/>

### 5-3 Set Style

<hr/>


#### Set color

- color 속성(attribute)을 이용해 색상을 정할 수 있음

````python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value + 100 for value in x2]

plt.plot(x1, y1, color="#000000")
plt.plot(x2, y2, c="c")

plt.show()
````

<br/>

#### Set linestyle

- ls 또른 linestyle 속성 사용

````python
plt.plot(x1, y1, c="b", linestyle="dashed")
plt.plot(x2, y2, c="r", ls="dotted")

plt.show()
````

<br/>

#### Set title

- pyplot에 title 함수를 사용한다.
- figure의 subplot 별로 title을 붙일 수 있다.

````python
plt.plot(x1, y1, c="b", linestyle="dashed")
plt.plot(x2, y2, c="r", ls="dotted")
plt.title("Two lines")

plt.show()
````

````python
# Latex 타입의 표현도 가능하다 (수식 표현)
plt.plot(x1, y1, c="b", linestyle="dashed")
plt.plot(x2, y2, c="r", ls="dotted")
plt.title('$y = \\frac{ax + b}{test}$')

plt.show()
````

<br/>

#### Set legend

- Legend 함수로 범례를 표시함, loc 위치 등 속성을 지정

````python
plt.plot(x1, y1, color="b", linestyle="dashed", label='line_1')
plt.plot(x2, y2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=False, loc="upper right")

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')


plt.show()
````

<br/>

#### Set grid & xylim

- Graph 보조선을 긋는 grid
- xy축 범위 한계를 지정하는 xylim

````python
plt.plot(x1, y1, color="b", linestyle="dashed", label='line_1')
plt.plot(x2, y2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")


plt.grid(True, lw=0.4, ls="--", c=".90")
plt.xlim(-100, 200)
plt.ylim(-200, 200)

plt.show()
````

### 5-4 Graph

<hr/>

#### Scatter

- 산점도
- marker 속성으로 모양을 지정할 수 있음

````python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:,0], data_1[:,1], c="b", marker="x")
plt.scatter(data_2[:,0], data_2[:,1], c="r", marker="o")

plt.show()
````

<br/>

- **s** 속성을 이용해 데이터의 크기를 지정할 수 있음

````python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
````

<br/>

#### Bar Chart

- Bar 함수 사용

````python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

x = np.arange(0,8,2)

plt.bar(x + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(x + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(x + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(x+0.50, ("A","B","C", "D"))
plt.show()
````

````python
data = np.array([[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
x = np.arange(data.shape[1])

for i in range(3):
    plt.bar(x, data[i], bottom = np.sum(data[:i], axis=0), 
            color = color_list[i], label=data_label[i])
plt.legend()
plt.show()
````

````python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
````

### 5-5 Matplotlib with Pandas

<hr/>

- Pandas 0.7 버전부터 matplotlib을 사용한 그래프를 지원
- DataFrame, Series별로 그래프 작성 가능

````python
import pandas as pd

# 보스턴 하우징 데이터
df_data = pd.read_csv("data/boston.csv")
df_data.head()
````

<br/>

- 데이터 간의 상관관계를 보기 위해 scatter graph를 사용

````python
fig = plt.figure()
ax = []
for i in range(1, 5):
    ax.append(fig.add_subplot(2,2,i))
ax[0].scatter(df_data["CRIM"], df_data["MEDV"], color="r", label="CRIM")
ax[1].scatter(df_data["PTRATIO"], df_data["MEDV"])
ax[2].scatter(df_data["AGE"], df_data["MEDV"])
ax[3].scatter(df_data["NOX"], df_data["MEDV"])
plt.show()
````

````python
fig = plt.figure()
fig.set_size_inches(10,5)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(df_data["MEDV"])
ax2.hist(df_data["MEDV"], bins=50)
ax1.set_title("House Price MEDV")
ax2.set_title("House Price MEDV")
````

<br/>

#### Scaled Boxplot

````python
# sklearn 으로 데이터 가져오는법
# from sklearn.datasets import load_boston

# boston_dataset = load_boston()
# df_data = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
````

````python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
scale_data = std_scaler.fit_transform(df_data)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.boxplot(scale_data, labels=df_data.columns)
fig.set_size_inches(18.5, 10.5)
````

<br/>

#### Scatter Matrix

````python
# pandas 자체가 가지고 있는 scatter_matrix 메소드
pd.plotting.scatter_matrix(df_data, diagonal="kde", alpha=1, figsize=(20,20))
plt.show()
````

````python
corr_data = np.corrcoef(scale_data.T)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
cax = ax.matshow(corr_data, vmin=-1, vmax=1, interpolation="nearest")
fig.colorbar(cax)
fig.set_size_inches(10,10)
ticks = np.arange(0,15,1)

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(df_data.columns)
ax.set_yticklabels(df_data.columns)
````
